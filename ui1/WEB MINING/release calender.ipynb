```python
from bs4 import BeautifulSoup
import requests
```

```python

url = "https://www.imdb.com/calendar/?ref_=nv_mv_cal"
page = requests.get(url)
if page.status_code==200:
    print("page downloaded")
else:
    print("not downloaded")
```

```python
soup = BeautifulSoup(page.text,'lxml')
```

```python
print(soup)
```

```python
soup.find('div',{'id':'main'})
```

```python
soup.find('h1')
soup.find_all('h4')
```

```python
main=soup.find('div',attrs={'id':'main'})
```

```python
dates = main.find_all('h4')
```

```python
names = main.find_all('ul')
```

```python
print(len(dates),len(names))
```

```python
for d ,n in zip(dates,names):
    print(d.text,n.text)
```

```python
for d,namelist in zip(dates,names):
    for name in namelist.find_all('li'):
        print(d.text,name.text.strip())
```

```python
realeases = []
for d,namelist in zip(dates,names):
    for name in namelist.find_all('li'):
        realeases.append({
            "dates":d.text,
            "names":name.text.strip()
        })
```

```python
import pandas as pd
```

```python
pd.DataFrame(realeases).to_csv('imdb_releases.csv') #comma seperated values = csv
```
