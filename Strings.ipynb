```python
v = ' Life before death,hope before despair,journey before destination'
from string import ascii_lowercase
for letter in ascii_lowercase:
  occurs = v.lower().count(letter)
  if occurs:
    print(f'{letter} -> {occurs}')
```

```python
q = input("enter a number")
if q.isnumeric():
  num = int(q)
  print('number = ',num)
elif  q.isalnum():
  temp = ''
  for char in q:
    if char.isnumeric():
      temp += char
  if temp:
      num = int(temp)
      print("data is convrted to number",num)
  else:
      print("your data has not been converted to numbers")
```
